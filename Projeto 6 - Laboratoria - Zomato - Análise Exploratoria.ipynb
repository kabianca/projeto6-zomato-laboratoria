## <font color=red>Introdução

Bangalore, oficialmente Bengaluru, é a capital do estado indiano de Karnataka, no sudoeste da Índia. Conhecida por seus inúmeros parques, concentrações de lojas e, sobretudo, um paraíso gastronômico. Seus sentidos estarão em alta enquanto você visita os mercados movimentados, cheios de especiarias perfumadas e ingredientes exóticos, pratos icônicos como o _dosa_ - um lanche semelhante a uma panqueca amado pelos habitantes locais. A Índia pode ser conhecida pelo chá, mas em Bangalore o que não falta são ótimas opções de cerveja artesanal, cafés e bares, os quais oferecem pratos dos mais diferentes lugares, para todos os gostos e orçamentos. Mais importante ainda, ajudará novos restaurantes a decidir quais os melhores locais para abertura de um novo empreendimento, seja voltado a uma cozinha específica, com base no valor cobrado por seus pratos, ou outra característica específica ao seu negócio. #### <font color=red> Georreferenciamento dos dados

```python
from geopy.geocoders import Nominatim
from tqdm import tqdm 
from folium import plugins
```

```python
# Georreferenciando as localidades, pois esse mesmo método não funcionou com os endereços, onde a grande maioria retorna erro:

i = 0

def findGeocode(city):  
    try:
        geolocator = Nominatim(user_agent="biatopeira@gmail.com") 
        loc = geolocator.geocode(city)
        return loc
    except GeocoderTimedOut:
        i += 1
        if i == 5:
            return None
        return findGeocode(city)   
```

```python
# Localização por localidade:

location_lat_long = {}
locations_unq = df1.location.unique()
for loc in locations_unq:
    loc = loc + ", Bangalore"
    location = findGeocode(loc)
    if location is None:
        print("Não encontrado... ", loc)
        continue
    lat = location.latitude
    long = location.longitude
    location_lat_long[loc] = [lat, long]
    print("Salvo.. ", loc)
```

```python
df2 = df1[(df1['location']!="St. Marks Road") & (df1['location']!="Rammurthy Nagar") & (df1['location']!="Sadashiv Nagar")]
```

```python
# Dispor os dados de geolocalização no mapa, indicando a concentração de restaurantes por localidade:

heatmap_df = []
for indx, row in tqdm(df2.iterrows(), total=len(df2)):
    if row['location']+", Bangalore" in location_lat_long:
        df2.loc[indx, 'latitude'] = location_lat_long[row['location']+", Bangalore"][0]
        df2.loc[indx, 'longitude'] = location_lat_long[row['location']+", Bangalore"][1]
        heatmap_df.append([location_lat_long[row['location']+", Bangalore"][0], location_lat_long[row['location']+", Bangalore"][1]])
```

```python
bengaluru = folium.Map(location=[12.972442, 77.580643], tiles='cartodbpositron', zoom_start=12.3)
```

```python
plugins.HeatMap(heatmap_df).add_to(bengaluru)
```

```python
# Remover números e indicadores de números ordinais da nomenclatura das localidades (esse fator irá agrupar localidades antes tidas 
# como distintas, tais como koramangala por exemplo:

df2['location_pr'] = df2.location.apply(lambda x:re.sub("[0-9]+(st|th|nd|rd) block", "", x.lower()).strip())
```

```python
# Criando uma variável para as localidades com maior número de restaurantes:

koramangala_btm = df2[(df2['location_pr'] == 'btm') | (df2['location_pr'] == 'koramangala')]
```

```python
# Plotagem dos restaurante melhor avaliados em Koramangala e BTM, dados que estas localidades concentram um número expressivo de restaurantes:

kora = koramangala_btm[koramangala_btm['location_pr'] == 'koramangala']
btm = koramangala_btm[koramangala_btm['location_pr'] == 'btm']

kora_rate_grp = kora.groupby(['rate']).agg({'name':'count'}).reset_index()
kora_rate_grp.sort_values(by="name", inplace=True, ascending=False)
btm_rate_grp = btm.groupby(['rate']).agg({'name':'count'}).reset_index()
btm_rate_grp.sort_values(by="name", inplace=True, ascending=False)

fig = plt.figure(figsize = (5, 4))
ax1 = plt.subplot(2,1,1)
plt.title('Top 5 restaurantes melhor avaliados de Koramangala versus BTM')
sns.barplot(x="rate", y="name", data=kora_rate_grp[:5], ax=ax1, palette='Reds_r')
plt.ylabel('Contagem (kor)')
ax2 = plt.subplot(2,1,2)
sns.barplot(x="rate", y="name", data=btm_rate_grp[:5], ax=ax2, palette='Reds_r')
plt.ylabel('Contagem (btm)')
plt.show()
```

```python
# Agrupar os restaurantes por localidade e avaliação para descobrir em quais localidades estão os restaurante melhor avaliados:

loc_rate_grp = df2.groupby(['location_pr', 'rate']).agg({'name':'count'}).reset_index()
loc_rate_grp.sort_values(by=["location_pr", 'name'], inplace=True, ascending=False)
loc_unq = df2.location_pr.unique()
for i, loc in enumerate(loc_unq):
    data = loc_rate_grp[(loc_rate_grp['location_pr'] == loc)]
    data = data[:5]
    if data[data.rate > 4].shape[0] > 1:
        fig = plt.figure(figsize = (5, 2.5))
        sns.barplot(x="rate", y="name", data=data, palette='Reds_r')
        plt.ylabel(f'Contagem ({loc})')
        plt.show()
```

```python
fig9 = plt.figure(figsize = (20, 10))
sns.boxplot(x="cost_for_two_people", y="location_pr", data=df2, ax=ax2, palette='Reds_r')
plt.show()
```

```python
# Análise dos restaurantes com maior tarifa nas localidades Lavelle Road e Church Street:

lavelle_church = df2[(df2['location_pr'] == 'lavelle road') | (df2['location_pr'] == 'church street')]
fig = plt.figure(figsize = (10, 5))
ax1 = plt.subplot(2,1,1)
plt.title('Comparação de preços entre restaurantes de Bangalore (Lavelle Road & Church Street)')
sns.boxplot(x="cost_for_two_people", data=df2, ax=ax1, palette='Reds_r')
ax2 = plt.subplot(2,1,2)
sns.boxplot(x="cost_for_two_people", y="location_pr", data=lavelle_church, ax=ax2, palette='Reds_r')
plt.show()
```

```python
# Onde estão localizados os restaurantes com o valor mais alto? Para isso a análise foi direcionada apenas as localidades onde o custo da refeição 
# para duas pessoas é maior que 2000:

fig = plt.figure(figsize = (20, 10))
expensive_restaurants = df2[df2.cost_for_two_people > 2000]
sns.boxplot(x='cost_for_two_people', y='location_pr', data=expensive_restaurants, palette='Reds_r')
```

```python
# Análise das localidades que concentram o maior número de restaurantes caros, sankey road, old airport road e residency road

super_expensive_res = df2[(df2['location_pr'] == 'sankey road') | (df2['location_pr'] == 'old airport road') | (df2['location_pr'] == 'residency road')]
super_expensive_res = super_expensive_res.groupby(['location_pr', 'rest_type']).agg({'name':'count'}).reset_index()
super_expensive_res.rename(columns={'name':'count'}, inplace=True)

# .barplot(data=super_expensive_res, x='count', y='location_pr', hue='rest_type')
fig, ax = plt.subplots(nrows=3, ncols=1, figsize = (10, 17))
sns.barplot(data=super_expensive_res[super_expensive_res.location_pr=='old airport road'], x='count', y='rest_type', ax=ax[0], palette='Reds_r')
ax[0].set_xlabel('Nº de restaurantes na old airport road')
sns.barplot(data=super_expensive_res[super_expensive_res.location_pr=='sankey road'], x='count', y='rest_type', ax=ax[1], palette='Reds_r')
ax[1].set_xlabel('Nº de restaurantes na sankey road')
sns.barplot(data=super_expensive_res[super_expensive_res.location_pr=='residency road'], x='count', y='rest_type', ax=ax[2], palette='Reds_r')
ax[2].set_xlabel('Nº de restaurantes na residency road')
```
